```python
df[bool_series]
```

```python
df[df['total_bill'] > 40]
```

```python
df[df['sex'] == "Male"]
```

```python
df[df['size'] > 3]
```

## Mutliple conditions

```python
df[(df['total_bill'] > 30) & (df['sex'] == "Male")]
```

```python
df[(df['total_bill'] > 30) | (df['sex'] == "Male")]
```

## Filtering by list of options

```python
df[(df['day'] == 'Sun') | (df['day'] == 'Sat')]
```

```python
options = ['Sat', 'Sun']
```

```python
df['day'].isin(options)
```

```python
df[df['day'].isin(options)]
```

# Useful Methods

```python
def last_four(num):
    return str(num)[-4:]
```

```python
df['CC Number'].apply(last_four)
```

```python
df['last_four'] = df['CC Number'].apply(last_four)
```

```python
df.head()
```

```python
df['total_bill']
```

```python
df['total_bill'].mean()
```

```python
def yelp(price):
    if price < 10:
        return '$'
    elif price > 10 and price < 30:
        return '$$'
    else:
        return '$$$'
```

```python
df['yelp'] = df['total_bill'].apply(yelp)
```
